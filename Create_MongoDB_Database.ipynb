```python
import pymongo
import pandas as pd
from tqdm import tqdm
data_csv = pd.read_csv('Data/Iris.csv')
myclient = pymongo.MongoClient("mongodb://localhost:27017/")  #Enter your MongoClient address
mydb = myclient["iris-data"]
mycol = mydb["Iris"]

for Id, SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm, Species in tqdm(data_csv.values):
    mylist = { "_id": Id, "SepalLengthCm": SepalLengthCm, "SepalWidthCm": SepalWidthCm, "PetalLengthCm": PetalLengthCm, "PetalWidthCm": PetalWidthCm, "Species": Species}
    dataset = mycol.insert_one(mylist)
```
